```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)
y = 1/np.cos(x)
z = 1/np.sin(x)
plt.plot(y, z, marker='x')
plt.plot(x, y, marker='x')
plt.plot(x, z, marker='x')
plt.plot(z, y, marker='x')
plt.plot(y, x, marker='x')
plt.plot(z, x, marker='x')
plt.show()
```
